```python
#Question 1
#A weather station records temperatures in C as

import numpy as np
data = [22.5 , 23.1 , 24.8 , 21.2 , 19.9]
array = np.array(data)
print("1. \nShape of array:\t", np.shape(array),"\n")
print("2. \nMean temperature:\t", np.mean(array),"\n")
sorted_array = np.sort(array)
print("3. \nSorted array:\t", sorted_array,"\n")
```

```python
#Question 2
#You have two quarterly sales arrays

import numpy as np
Q1 = np.array([500 , 600 , 700])
Q2 = np.array([450 , 550 , 800])

concarray = np.vstack((Q1, Q2))
print("1. \nConcatenated array:\n",concarray,"\n")
splitarray = np.split(concarray,1)
print("2. \nSplit array:\n",splitarray,"\n")
```

```python
#Question 3
#Given a NumPy array of stock prices:

import numpy as np
prices = np.array([120 , 122 , 121 , 125 , 123 , 124])
reshapedarray = np.reshape(prices, (3,2))
horz_split = np.hsplit(reshapedarray,2)
deep_copy = horz_split[0]
print(deep_copy)
```

```python
#Question 4
#Given a 2D dataset:

import numpy as np
X = np.array([[1 , 2] , [3 , 4] , [5 , 6]])
cov_X = np.cov(X.T)
print("1. \nCovariance Matric",cov_X,"\n")
eig_val,eig_vec = np.linalg.eig(cov_X)
print("2. \nEigenvalues",eig_val,"\n")
print("3. \nEigenvectors",eig_vec,"\n")
```

```python
#Question 5
#Simulate 1000 trials of stock returns assuming:

normal_returns = np.random.normal(loc=0.05, scale=0.1, size=1000)

print("Simulation of Stock Returns\n")
print(normal_returns,"\n\n")

posson_returns = np.random.poisson(lam=3, size=1000)

print("Simulation with Posson Distribution\n")
print(posson_returns)
```
