```python
#Q1: Take out employee dataset and perform calculations
import pandas as pd
import matplotlib.pyplot as plt
data_emp = {
    "Name": ["EmpA","EmpB","EmpC","EmpD","EmpE","EmpF","EmpG","EmpH"],
    "Salary": [72000,85000,64000,95000,78000,88000,67000,91000],
    "Department": ["HR","Engineer","Sales","Tech","Sales","Engineer","HR","Tech"],
    "Years_Worked": [3,7,2,10,5,8,4,9]
}

df1 = pd.DataFrame(data_emp)
display(df1)

dep_avg = df1.groupby("Department")["Salary"].mean()
print("\n\nAverage Salary by Department")
display(dep_avg)

df1["Bonus"] = df1["Salary"] * (df1["Years_Worked"] / 10)
print("\n\nBonus of Employees calculated by years worked")
display(df1)

df1["Annual Salary"] = (df1["Salary"]*12)+df1["Bonus"]
print("\n\nAnnual Salary of Employees")
display(df1)

print("\n\nEMPLLOYEEE STATICTICS\n")
avg = df1["Salary"].mean()
print("\nAverage Salary of all employees: ",avg)
max = df1["Salary"].max()
min = df1["Salary"].min()
print("Max Salary: ",max)
print("Min Salary: ",min)
total_salary_paid = df1["Annual Salary"].sum()
print("Annual Salary paid to all employeed: ",total_salary_paid)

print("\n\n")
plt.boxplot(df1["Salary"])
plt.title("Salary Box Plot")
plt.show()
```

```python
#Q2: Take a student dataset and analyze grades of the subject
import matplotlib.pyplot as plt
data_stud = {
    "Name": ["StudA", "StudB", "StudC", "StudD", "StudE", "StudF", "StudG", "StudH"],
    "Maths": [88, 76, 92, 81, 67, 95, 73, 89],
    "Physics": [84, 83, 91, 77, 72, 94, 69, 85],
    "Chemistry": [90, 74, 88, 80, 70, 96, 75, 83]
}
df2 = pd.DataFrame(data_stud)
display(df2)

df2["Average"] = (df2["Maths"]+df2["Physics"]+df2["Chemistry"])/3
print("\n\nAverage Marks")
display(df2)

print("\n\nStudent Statistics")
max_average = df2["Average"].max()
min_average = df2["Average"].min()
print("\nStudent with max average: ",df2.loc[df2["Average"]==max_average,"Name"].iloc[0])
print("\nStudent with min average: ",df2.loc[df2["Average"]==min_average,"Name"].iloc[0])


print("\n\n")
df2.plot(x="Name",y="Average",kind="bar")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.title("Average Marks of Students")
plt.xticks(rotation=45)
plt.show()

print("\n\n")
df2.plot.bar(x="Name",y=["Maths","Physics","Chemistry"])
plt.xlabel("Name")
plt.ylabel("Marks")
plt.title("Marks of Students in all the Subjects")
plt.xticks(rotation=45)
plt.show()

print("\n\nStudent Subject Statistics: \n")
math_max = df2["Maths"].max()
phys_max = df2["Physics"].max()
chem_max = df2["Chemistry"].max()
math_min = df2["Maths"].min()
phys_min = df2["Physics"].min()
chem_min = df2["Chemistry"].min()
print("Student with max maths: ",df2.loc[df2["Maths"]==math_max,"Name"].iloc[0])
print("Student with min maths: ",df2.loc[df2["Maths"]==math_min,"Name"].iloc[0])
print("Student with max physics: ",df2.loc[df2["Physics"]==phys_max,"Name"].iloc[0])
print("Student with max physics: ",df2.loc[df2["Physics"]==phys_min,"Name"].iloc[0])
print("Student with max chemistry: ",df2.loc[df2["Chemistry"]==chem_max,"Name"].iloc[0])
print("Student with max chemistry: ",df2.loc[df2["Chemistry"]==chem_min,"Name"].iloc[0])

print("\n\nSubject Statistics: \n")
subj_avg = df2[["Maths","Physics","Chemistry"]].mean()
display(subj_avg)
subj_avg_max = subj_avg.max()
subj_avg_min = subj_avg.min()
print("Subject with max average: ",subj_avg.idxmax())
print("Subject with min average: ",subj_avg.idxmin())
```
