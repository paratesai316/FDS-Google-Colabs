```python
#question 1
import sqlite3
con = sqlite3.connect("company.db")
cur = con.cursor()
cur.execute("DROP TABLE IF EXISTS employees")
cur.execute("CREATE TABLE IF NOT EXISTS employees (id INTEGER PRIMARY KEY AUTOINCREMENT, name TEXT, dept TEXT)")
def add_employee(name, department):
  cur.execute("INSERT INTO employees (name, dept) VALUES (?, ?)", (name, department))
  con.commit()
add_employee("Sai", "DSBS")
add_employee("Parate", "C Tech")
cur.execute("SELECT * FROM employees")
print(cur.fetchall())
con.close()
```
